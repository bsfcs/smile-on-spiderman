```python
visits = smile_eda['CallerNum'].value_counts()
visits
```

```python
patients = visits.to_frame().reset_index()
patients.head()
```

```python
patients = patients.rename(columns = {'index': 'caller_num', 'CallerNum': 'number_of_visits'})
patients.head()                           
```

```python
patients['success'] = ''
patients.head()
```

```python
for ind, row in smile_eda.iterrows():
    if row['treatment_plan_completed?'] == 'Yes':
        patients.loc[ind, 'success'] = 'Yes'
    elif row['treatment_plan_completed?'] == 'No':
        patients.loc[ind, 'success'] ='No'
    else:
        patients.loc[ind, 'success'] ='unknown'
patients.head(25)
```
