```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
smile_on = pd.read_csv('../data/smile_on.csv', encoding="ISO-8859-1",  low_memory=False, header=2)
#for line in open('u.item', encoding='ISO-8859-1'):
#read each line
smile_on.head()
```

```python
smile_on = smile_on.rename(columns = {'SMILE ON 60+ Screening - Last Screening Date': 'Last_Screen_Date',
                                      'SMILE ON 60+ Case Management Information - Dental Clinic' : 'Case_Mgmt_Clinic',
                                      'SMILE ON 60+ Clinic Information - Is patient of record date if yes last seen?' :'Date_Last_Seen',
                                      'SMILE ON 60+ Clinic Information - Was initial appointment made?' :'Initial_Appt_Made',
                                      'SMILE ON 60+ Demographic Information - When is the last time you visited your dentist?': 'Last_Dentist_Visit',
                                      'SMILE ON 60+ Oral Care Encounter - What care was provided to the enrollee? (Select all that apply)' : 'Care_Provided',
                                      'SMILE ON 60+ Screening - How often do you brush your teeth?' : 'Brush_Frequency',
                                      'SMILE ON 60+ Screening - Need for Periodontal Care' : 'Need_For_Care',
                                      'SMILE ON 60+ Screening - Root Fragments' : 'Root_Fragments',
                                      'SMILE ON 60+ Screening - Suspicious Soft Tissue Lesions' : 'Lesions',
                                      'SMILE ON 60+ Screening - Treatment Urgency' : 'Urgency',
                                      'SMILE ON 60+ Screening - Untreated Decay' : 'Decay',
                                      'SMILE ON 60+ Site Information - Enroller ID' : 'Clinic_Attended',
                                      'SMILE ON 60+ Treatment Plan - Is the treatment plan completed?' : 'Tx_Plan_Complete',
                                      'SMILE ON 60+ Treatment Plan - Was a treatment plan developed?' : 'Tx_Plan_Developed',
                                      'SMILE ON 60+ Treatment Plan - Was the enrollee able to be restored to function (can chew) and "social six" esthetics (top front six teeth are present and disease free)' : 'Function_Restored',
                                      'SMILE ON 60+ Base-line Oral and Conditions Questions - Are you limited in what you can eat?': 'limited_eating',
                                      'SMILE ON 60+ Base-line Oral and Conditions Questions - Dental Clinic': 'dental_clinic',
                                      'SMILE ON 60+ Site Information - Enroller Location': 'enroller_location',
                                      'SMILE ON 60+ Treatment Plan - can chew and "social six" esthetics': 'social_six',
                                      'SMILE ON 60+ Verification of Eligibility - Transportation Arranged': 'transportation_arranged'
                                       })
smile_on.head()
```

```python
#change time format
smile_on['CallDateAndTimeStart'] = smile_on['CallDateAndTimeStart'].str.split(' ').apply(lambda x: str(x[0]))
```

```python
visits = smile_on['CallerNum'].value_counts()
```

```python
#create dataframe based on unique callers
patients = visits.to_frame().reset_index()
patients.head()
```

```python
#fix columns
patients = patients.rename(columns = {'index': 'caller_num', 'CallerNum': 'number_of_visits'})
patients.head() 
```

```python
#remove unique callers -1,-2
patients = patients.loc[~patients['caller_num'].isin(['-1', '-2'])]
patients.shape
```

```python
#add time column
patients['start_date'] = ''

for ind, row in smile_on.iterrows():
    patients.loc[ind, 'start_date'] = row['CallDateAndTimeStart']
patients.head()
```

```python
#add care_provided column
patients['care_provided'] = ''

for ind, row in smile_on.iterrows():
    patients.loc[ind, 'care_provided'] = row['Care_Provided']
patients.head()
```

```python
#add treatment plan column
patients['treatment_plan'] = ''

for ind, row in smile_on.iterrows():
    if row['Tx_Plan_Developed'] == 'Yes':
        patients.loc[ind, 'treatment_plan'] = ['treatment_plan'].count('Yes') + 1
    elif row['Tx_Plan_Developed'] == 'No':
        patients.loc[ind, 'treatment_plan'] ='no'
    else:
        patients.loc[ind, 'treatment_plan'] ='unknown'
patients.head()
```

```python
#add treatment success column
patients['success'] = ''

for ind, row in smile_on.iterrows():
    if row['Tx_Plan_Complete'] == 'Yes':
        patients.loc[ind, 'success'] = ['success'].count('Yes') + 1
    elif row['Tx_Plan_Complete'] == 'No':
        patients.loc[ind, 'success'] ='no'
    else:
        patients.loc[ind, 'success'] ='unknown'
patients.head(25)
```

```python
#add social_six column
patients['social_six'] = ''

for ind, row in smile_on.iterrows():
    if row['social_six'] == 'Yes':
        patients.loc[ind, 'social_six'] = 'yes'
    elif row['social_six'] == 'No':
        patients.loc[ind, 'social_six'] ='no'
    else:
        patients.loc[ind, 'social_six'] ='unknown'
patients.head()
```

```python
#add postal code column
patients['postal_code'] = ''

for ind, row in smile_on.iterrows():
    patients.loc[ind, 'postal_code'] = row['PostalCode']
patients.head()
```

```python
#add clinic column
patients['clinic_attended'] = ''

for ind, row in smile_on.iterrows():
    patients.loc[ind, 'clinic_attended'] = row['Clinic_Attended']
patients.head()
```
