```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
cms_payments = pd.read_csv('../data/cms_physician_and_other_supplier-payments_2015-2017.csv')
#for line in open('u.item', encoding='ISO-8859-1'):
#read each line
cms_payments.columns
```

```python
cms_payments.head()
```

```python
#df.loc[(df['First_name'] == 'Bill') | (df['First_name'] == 'Emma')
cms_payments=cms_payments.loc[cms_payments['year'].isin([2016,2017])] #df[df.Name.isin(['Alice', 'Bob'])]
#cms_payments = cms_payments.loc[cms_payments['year'] == 2016] | cms_payments.loc[cms_payments['year'] == 2017]
#cms_payments.head()
cms_payments.head()
```

```python
cms_payments = cms_payments[['City of the Provider',
                             'Average Submitted Charge Amount',
                             'Average Medicare Payment Amount']].copy()
cms_payments
```

```python
cms_payments = cms_payments.rename(columns = {'City of the Provider':'city',
                                                'Average Submitted Charge Amount':'avg_submitted',
                                                'Average Medicare Payment Amount':'avg_medicare_payment'
                                               })
```

```python
cms_payments['percent_covered']=(cms_payments['avg_medicare_payment']/cms_payments['avg_submitted'])*100
cms_payments.head()
```

```python
cms_payments=cms_payments.groupby(['city']).mean().reset_index()
cms_payments.info(25)
```

```python
cms_payments['avg_submitted']=cms_payments['avg_submitted'].round(2)
cms_payments['avg_medicare_payment']=cms_payments['avg_medicare_payment'].round(2)
cms_payments['percent_covered']=cms_payments['percent_covered'].round(2)
cms_payments.head(15)
```

```python
cms_payments['city']=cms_payments['city'].str.title()
cms_payments
```

```python
#export to csv
cms_payments.to_csv('../cms_payments.csv', index = False)
```
