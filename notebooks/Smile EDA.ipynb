```python
visits = smile_on['CallerNum'].value_counts()
```

```python
#create dataframe based on unique callers
patients = visits.to_frame().reset_index()
patients.head()
```

```python
#fix columns
patients = patients.rename(columns = {'index': 'caller_num', 'CallerNum': 'number_of_visits'})
patients.head()
```

```python
#remove unique callers -1,-2
patients = patients.loc[~patients['caller_num'].isin(['-1', '-2'])]
patients.shape
```

```python
#add time column
patients['start_date'] = ''

for ind, row in smile_on.iterrows():
    patients.loc[ind, 'start_date'] = row['CallDateAndTimeStart']
patients.head()
```

```python
#add care_provided column
patients['care_provided'] = ''

for ind, row in smile_on.iterrows():
    patients.loc[ind, 'care_provided'] = row['Care_Provided']
patients.head()
```

```python
#add treatment plan column
patients['treatment_plan'] = ''

for ind, row in smile_on.iterrows():
    if row['Tx_Plan_Developed'] == 'Yes':
        patients.loc[ind, 'treatment_plan'] = ['treatment_plan'].count('Yes') + 1
    elif row['Tx_Plan_Developed'] == 'No':
        patients.loc[ind, 'treatment_plan'] ='no'
    else:
        patients.loc[ind, 'treatment_plan'] ='unknown'
patients.head()
```

```python
#add treatment success column
patients['success'] = ''

for ind, row in smile_on.iterrows():
    if row['Tx_Plan_Complete'] == 'Yes':
        patients.loc[ind, 'success'] = ['success'].count('Yes') + 1
    elif row['Tx_Plan_Complete'] == 'No':
        patients.loc[ind, 'success'] ='no'
    else:
        patients.loc[ind, 'success'] ='unknown'
patients.head(25)
```

```python
#add postal code column
patients['postal_code'] = ''

for ind, row in smile_on.iterrows():
    patients.loc[ind, 'postal_code'] = row['PostalCode']
patients.head()
```

```python
#add clinic column
patients['clinic_attended'] = ''

for ind, row in smile_on.iterrows():
    patients.loc[ind, 'clinic_attended'] = row['Clinic_Attended']
patients.head()
```

```python
#add last visit column
patients['last_visit'] = ''

for ind, row in smile_on.iterrows():
    patients.loc[ind, 'last_visit'] = row['Date_Last_Seen']
patients.head()

```

```python
#Create new csv with narrowed down data
patients_df = patients.to_csv('../data/patients.csv')
```
